```python
def RGB2YUV( rgb ):
    # the jpeg standard uses full 8 bit with no reserved values
    m = np.array([[ 0.299, 0.587,  0.114],
                 [-0.168736, -0.331264, 0.5],
                 [ 0.5, -0.418688, -0.08131]])
    yuv = np.dot(rgb, m.transpose()) + np.array([0, 128, 128])
    # referred from page 98, SzeliskiBookDraft_20210214
    return yuv

img = Image.open('Clab assignment 1 images/Fig2a.png')
img_yuv = RGB2YUV(img.convert("RGB")) # converting the image to jpg my function is designed to work on jpeg image
y, u, v = cv2.split(img_yuv)


#showing the original image in rgb
plt.subplot(1, 5, 1).set_title('rgb')
# explicitly casting image to uint8 before displaying it.
plt.imshow(np.uint8(img)) 



#showing the converted yuv image
plt.subplot(1, 5, 2).set_title('yuv')
# explicitly casting image to uint8 before displaying it.
plt.imshow(np.uint8(img_yuv)) 


#showing the y channel
plt.subplot(1, 5, 3).set_title('y channel')
# explicitly casting image to uint8 before displaying it.
plt.imshow(np.uint8(y)) 

#showing the u channel
plt.subplot(1, 5, 4).set_title('u channel')
# explicitly casting image to uint8 before displaying it.
plt.imshow(np.uint8(u))

#showing the v channel
plt.subplot(1, 5, 5).set_title('v channel')
# explicitly casting image to uint8 before displaying it.
plt.imshow(np.uint8(v))
plt.show()
```

2. Implement your own Matlab/python function that performs a 5x5 Gaussian filtering (1.5 marks). Your function interface is:
my_Gauss_filter()
input: noisy_image, my 11*11 gausskernel
output: output_image

```python
def gaussian_filter(image, k_size, sigma):
    height, width = image.shape[0], image.shape[1]
    # distance image height and width
    distnce_height = height - k_size + 1
    distnce_width = width - k_size + 1
    # adding 1 above since indexing is from 0
    # distance width and height is calculated to form image_array
    image_array = np.zeros((distnce_height * distnce_width, k_size * k_size))
    row = 0
    # using product imprted from itertools
    for i, j in product(range(distnce_height), range(distnce_width)):
        window = np.ravel(image[i : i + k_size, j : j + k_size])
        image_array[row, :] = window
        row = row + 1

    gaussian_kernel = gen_gaussian_kernel(k_size, sigma)
    # by calling the function above we get the kernel matrix
    filter_array = np.ravel(gaussian_kernel)

    tmp = np.dot(image_array, filter_array)
    # reshaping and storing as unit 8 type
    distnce = tmp.reshape(distnce_height, distnce_width).astype(np.uint8)

    return distnce

# helper function for gaussian filter
def gen_gaussian_kernel(k_size, sigma):
    # function to generate gaussian kernel of given kernel size
    center = k_size // 2
    # Above we divide kernel size by 2 as floor division to find the center
    x, y = np.mgrid[0 - center : k_size - center, 0 - center : k_size - center]
    mat_g = 1 / (2 * np.pi * sigma) * np.exp(-(np.square(x) + np.square(y)) / (2 * np.square(sigma)))
    return mat_g
```

# Image Denoising via a Bilateral filter

1. Using your Gaussian filter as a base, implement your own Matlab/Python function that performs a 5x5 Bilateral filtering to gray-scale image. (1.5 marks) Your function interface must be:
my_Bilateral_filter()
input: noisy_image, my_11x11_gausskernel, colour_sigma
output: output_image
where colour_sigma is the sigma applied to the range (intensity/gray scale) part
of the filter.

```python
def slice_img(imege, x, y, kernel_size):
    # slicing the image
    # floor division of kernel size
    half = kernel_size // 2
    return imege[x - half : x + half + 1, y - half : y + half + 1]


def gauss_kernel_find(kernel_size, spatial_variance):
    # this functions creates the gaussian kernel
    arr = np.zeros((kernel_size, kernel_size))
    for i in range(0, kernel_size):
        for j in range(0, kernel_size):
            arr[i, j] = np.sqrt(
                abs(i - kernel_size // 2) ** 2 + abs(j - kernel_size // 2) ** 2
            )
    return gauss_vec(arr, spatial_variance)

def gauss_vec(imege, var):
    # this function helps in applying gaussian function to each element in matrix
    sigma = np.sqrt(var)
    constant = 1 / (sigma * np.sqrt(2 * np.pi))
    return constant * np.exp(-((imege / sigma) ** 2) * 0.5)


def bilateral_filter(imege, spatial_variance, intensity_variance, kernel_size):
    # performs bilateral filter
    row_Y, col_Y = imege.shape
    imege2 = np.zeros(imege.shape)
    gaussKernel = gauss_kernel_find(kernel_size, spatial_variance)
    for i in range(kernel_size // 2, row_Y - kernel_size // 2):
        for j in range(kernel_size // 2, col_Y - kernel_size // 2):
            imegeS = slice_img(imege, i, j, kernel_size)
            imegeI = imegeS - imegeS[kernel_size // 2, kernel_size // 2]
            # calling and passing intensity variance to gauss_vec
            imegeIG = gauss_vec(imegeI, intensity_variance)
            wghts = np.multiply(gaussKernel, imegeIG)
            vals = np.multiply(imegeS, wghts)
            val = np.sum(vals) / np.sum(wghts)
            imege2[i, j] = val
    return imege2
```

2. Compare forward and backward mapping and analyze their difference (1.0 mark).

```python
def mapForw (img,Sx,Sy):
    # function for forward mapping
    # Way of declaring an empty image. The data type given is ubyte for image
    # since it is color image channel is set three
    Forward_img = np.ones((int(rows + cols*Sx),int(cols + rows*Sy),3), dtype=np.ubyte)
    # performing forward mapping
    for row in range(rows):
        for col in range(cols):
            Forward_img[int(row+col*Sx), int(col+row*Sy),:] = img[row,col,:]
    return Forward_img

def mapBack (img, Sx, Sy):
    # function for backward mapping
    # Way of declaring an empty image. The data type given is ubyte for image
    # since it is color image channel is set three
    Backward_img= np.ones(shape=img.shape, dtype=np.ubyte);
    # performing backward mapping
    for row in range(rows):
        for col in range(cols):
            backCol = int (col-row*Sy)
            backRow = int (row-col*Sx)
            Backward_img[backRow, backCol, :] = img[row,col,:]
    return Backward_img

img = imageio.imread('Clab assignment 1 images/image1.jpg')
img_resized = cv2.resize(img,(512,512), interpolation=cv2.INTER_AREA)
rows, cols, _ = img_resized.shape

Sx = 0.12 # Scaling size for width
Sy = 0.14 # Scaling size for height

fimg = mapForw(img_resized, Sx, Sy)
bimg = mapBack(fimg, Sx, Sy)
bimg_org = mapBack(img_resized, Sx, Sy)

plt.subplot(1, 4, 1).set_title('Original')
plt.imshow(np.uint8(img_resized))

plt.subplot(1, 4, 2).set_title('Forward Map')
plt.imshow(np.uint8(fimg))

plt.subplot(1, 4, 3).set_title('BMap on FMap')
plt.imshow(np.uint8(bimg))

plt.subplot(1, 4, 4).set_title('Backward Map')
plt.imshow(np.uint8(bimg_org))

plt.show()
```
